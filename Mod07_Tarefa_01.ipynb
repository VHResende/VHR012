````python
# Criando um dataframe com os nomes de cada variável e o tipo de dados de cada variável:
variaveis = pd.DataFrame({
    'variavel': df.columns,
    'tipo_dado': df.dtypes
})

# Adicionando uma coluna 'qtd_categorias':
qtd_categorias = []

# Contando o número de categorias para cada variável:
for coluna in df.columns:
    qtd_categorias.append(df[coluna].nunique())

# Adicionando a lista 'qtd_categorias' ao dataframe:
variaveis['qtd_categorias'] = qtd_categorias

# Exibindo o dataframe resultante:
print(variaveis)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# Identificandovariáveis qualitativas (categóricas):
categorical_vars = df.select_dtypes(include=['object']).columns
print("Variáveis categóricas:")
print(categorical_vars)

# Criando variáveis dummy para as variáveis categóricas:
df_dummies = pd.get_dummies(df, columns=categorical_vars, drop_first=True)

# Excluindo variáveis do tipo string (manter apenas suas versões dummy).
# Selecionando colunas numéricas e binárias.
numeric_and_binary_vars = df.select_dtypes(include=['number', 'bool']).columns
df_final = df_dummies[numeric_and_binary_vars.union(df_dummies.columns.difference(categorical_vars))]

# Exibindo o dataframe final
print("Dataframe final para entrada no scikit-learn:")
print(df_final.head())
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
# Tabela cruzada entre 'possui_email' e 'mau'
tabela_possui_email = pd.crosstab(df['possui_email'], df['mau'], normalize='index')
print("Tabela cruzada - possui_email vs mau:")
print(tabela_possui_email)

# Tabela cruzada entre 'posse_de_veiculo' e 'mau'
tabela_posse_veiculo = pd.crosstab(df['posse_de_veiculo'], df['mau'], normalize='index')
print("Tabela cruzada - posse_de_veiculo vs mau:")
print(tabela_posse_veiculo)
````

Análise:

Para a variável 'possui_email', a proporção de 'mau = 1' é maior quando 'possui_email = False'.
Para 'posse_de_veiculo', as proporções de 'mau = 1' são relativamente próximas entre as categorias.
Portanto, 'possui_email' parece ser mais poderosa para prever a probabilidade de 'mau = 1'.
